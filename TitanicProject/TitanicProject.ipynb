```python
"""1.使用逻辑回归算法进行分类和预测泰坦尼克号人员生存还是死亡"""
import numpy as np
from numpy import *
import pandas as pd
import csv

"""定义sigmoid函数"""
def sigmoid(inX):
    return 1.0/(1+exp(-inX))

"""以回归系数和特征向量作为输入计算预测的类别"""
def classify(inX,weights):
    prediction = sigmoid(sum(inX * weights))
    if prediction>0.5:
        return 1.0
    else:
        return 0.0

"""改善的随机梯度上升法"""
def StocGradAscent_improved_V1(dataMatrix,labelMatrix,numIter = 500):
    m,n = shape(dataMatrix)
    weights = ones(n)
    for j in range(numIter):
        dataIndex = list(range(m))
        for i in range(m):
            alpha = 4/(1.0+j+i) + 0.01   #动态更新alpha值,越来越小,但最小不会超过0.01
            randIndex = int(random.uniform(0,len(dataIndex)))
            prediction = sigmoid(sum(dataMatrix[randIndex] * weights))  #随机选取更新点,避免连续点引起的周期性抖动
            error = labelMatrix[randIndex] - prediction
            weights = weights + alpha * error * dataMatrix[randIndex]
            del(dataIndex[randIndex])
    return weights

"""对训练集和测试集进行处理"""
def dealdata():
    """加载训练集和测试集"""
#     trainSet = pd.read_csv('/home/jerry/GitRepository/竞赛/dataset/TitanicProjectData/train.csv',usecols=range(10))
#     testSet = pd.read_csv('/home/jerry/GitRepository/竞赛/dataset/TitanicProjectData/test.csv',usecols=range(10))
#     trainLabels =  pd.read_csv('/home/jerry/GitRepository/竞赛/dataset/TitanicProjectData/train.csv',usecols=[11])
#     testLabels = pd.read_csv('/home/jerry/GitRepository/竞赛/dataset/TitanicProjectData/test.csv',usecols=[11])
    csv_train = open('./data/train.csv')
    csv_test = open('./data/test.csv')
    trainSet = []
    trainLabels = []
    lineNum = 0  #统计加载数据的行数    
  
    for line in csv_train.readlines():
        currline = line.strip().split(',')
        lineArr = []  #一行一行添加到数据集中
        for i in currline[:12]:
            lineArr.append(i)
        trainSet.append(lineArr)
        trainLabels.append(currline[-1])
    print(shape(trainSet))
#     trainSet.dtype
#     print('trainSet:',trainSet)
    trainWeights = StocGradAscent_improved_V1(trainSet,trainLabels)  #计算训练集的梯度
    accCount = 0  #统计分类准确的个数
    numTest = 0 #统计纳入分类统计的总数
    
    """加载测试集,并对每一个加载的数据进行测试分类正确性"""   
    for line1 in csv_test.readlines():
        numTest = numTest + 1  #每次加载一行,计算参与分类数据+1
        currLine1 = line1.strip().split(',')
        lineArr1 = []
        for i in currline[:12]:
            lineArr1.append(i)    
        if classify(lineArr1,trainWeights) == currLine1[-1]:  #判断加载的每一个数据分类是否准确
            accCount = accCount + 1

    accuracy = (float(accCount) / numTest)
    print('the accuracy is : %s' % accuracy)       
    frtrain.close()
    frtest.close()
    return accuracy 

"""多次测试"""
def multiTest():
    numTest = 20
    accSum = 0
    """进行numTest数量测试,计算所有准确率之和"""
    for k in range(numTest):
        accSum = accSum + dealdata()
    print('after %d iteration,the average accuracy is:%s' % (numTest,accSum/float(numTest)))   

# """绘制分类结果的决策边界"""
# def plotBestFit()
    
if __name__ == '__main__':
    multiTest()   
```

```python
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract('([A-Za-z]+)\.',expand=False)
pd.crosstab(train_df['Title'], train_df['Sex'])  #找出每个name缩写的人数
```

```python
for dataset in combine:  #替换和融合原来特征,并且生成新的特征,然后按照某一特征进行分类
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
train_df[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

```python
title_mapping = {'Mr':1, 'Miss':2, 'Mrs':3, 'Rare':4}
for dataset in  combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)
train_df.head()  #将非数值型特征转换为数值型特征
```

```python
train_df = train_df.drop(['Ticket'], axis=1)
test_df = test_df.drop(['Name'], axis=1)
combine = [train_df, test_df]
train_df.shape, test_df.shape  #删除掉已经融合过得特征
```

```python
train_df.head()
```

```python
for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map({'female':1, 'male':0}).astype(int)
train_df.head()  #利用map函数直接将非数值型数值转换为数值型数据
```

```python
train_df['AgeBand'] = pd.cut(train_df['Age'], 5) #将年龄均分成5个年龄段
train_df[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
#计算各个年龄段的存活率,按照年龄段升序排序
train_df[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean().sort_values(by='Survived', ascending=True)
```

```python
#用序号来代替每个年龄段
for dataset in combine:    
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age']
train_df.head()
```

```python
#删除掉AgeBend特征
train_df = train_df.drop(['Cabin','AgeBand'], axis=1)
combine = [train_df, test_df]
train_df.head()
```

```python
#组合现有的特征,融合成新的特征
for dataset in combine:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1  #将SlibSp,Parch合成一列
train_df[['Survived', 'FamilySize']].groupby(['FamilySize'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
#融合新的特征 是否独自乘船
for dataset in combine:
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamilySize'] ==1, 'IsAlone'] = 1
train_df[['IsAlone', 'Survived']].groupby(['IsAlone'], as_index = False).mean()
```

```python
#将SibSp,Parch和FamilySize融合成一列IsAlone
train_df = train_df.drop(['SibSp', 'Parch', 'FamilySize'], axis=1)
test_df = test_df.drop(['SibSp', 'Parch', 'FamilySize'], axis=1)
combine = [train_df, test_df]
train_df.head(6)
```

```python
#直接对现有特征进行加减乘除生成新的特征
for dataset in combine:
    dataset['Age'] = dataset['Age'].fillna(0)
    dataset['Age*Pclass'] = dataset.Age * dataset.Pclass
train_df.loc[:,['Age*Pclass', 'Age', 'Pclass']].head(10)
```

```python
freq_port = train_df.Embarked.dropna().mode()[0]  #计算Embarked特征最高频率值
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].fillna(freq_port)  #用最高频率值替代缺失值
train_df[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
#把分类要素转换为数字
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].map({'S':0, 'C':1, 'Q':2}).astype(int)
train_df.head()
```

```python
test_df['Fare'].fillna(test_df['Fare'].dropna().median(), inplace=True)
test_df.head()
```

```python
train_df['FareBand'] = pd.qcut(train_df['Fare'], 4)
train_df[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=True)
```

```python
#将Fare按区间分成四段之后映射成4个数字
for dataset in combine:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)

train_df = train_df.drop(['FareBand'], axis=1)
combine = [train_df, test_df]
    
train_df.head(10)
```
