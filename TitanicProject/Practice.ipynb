```python
'''以下为数据预处理部分'''
import numpy as np
from numpy import *
import pandas as pd

train_df = pd.read_csv('./data/train.csv')
test_df = pd.read_csv('./data/test.csv')
combine = [train_df, test_df]
print(train_df.columns.values)
```

```python
train_df[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
train_df[['Sex', 'Survived']].groupby(['Sex'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
print('Before',train_df.shape, test_df.shape, combine[0].shape, combine[1].shape)
test_df = test_df.drop(['Ticket', 'Cabin'], axis=1)
train_df = train_df.drop(['Ticket', 'Cabin'], axis=1)
combine = [train_df, test_df]
print('After',train_df.shape, test_df.shape, combine[0].shape, combine[1].shape)
```

```python
train_df.info()
```

```python
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract('([A-Za-z]+)\.', expand=False)
pd.crosstab(train_df['Title'], train_df['Sex'])
```

```python
for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
train_df[['Title', 'Survived']].groupby(['Title'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
train_df.head()
```

```python
title_mapping = {'Mrs':1, 'Miss':2, 'Master':3, 'Rare':4, 'Mr':5}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)
train_df.head()
```

```python
train_df = train_df.drop(['Name'], axis=1)
train_df.head()
combine = [train_df, test_df]
```

```python
freq_sex = train_df.Sex.dropna().mode()[0]
for dataset in combine:
    dataset['Sex'] = dataset['Sex'].fillna(freq_sex)
    dataset['Sex'] = dataset['Sex'].map({'female':1, 'male':0}).astype(int)

train_df.head()  #利用map函数直接将非数值型数值转换为数值型数据
```

```python
train_df['AgeBand'] = pd.cut(train_df['Age'], 5) #将年龄均分成5个年龄段
train_df[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
#计算各个年龄段的存活率,按照年龄段升序排序

```

```python
#用序号来代替每个年龄段
for dataset in combine:    
    dataset.loc[ dataset['Age'] <= 16.0, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16.0) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32.0) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48.0) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64.0, 'Age']
test_df.head()
```

```python
train_df.head()
```

```python
train_df = train_df.drop(['AgeBand'], axis=1)
train_df.head()
```

```python
test_df = test_df.drop(['Name'], axis=1)
```

```python
test_df.head()
```

```python
combine = [train_df, test_df]
```

```python
for dataset in combine:
    dataset['FamileSize'] = dataset['SibSp']+dataset['Parch'] + 1 
```

```python
for dataset in combine:
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamileSize']==1, 'IsAlone'] = 1
```

```python
train_df = train_df.drop(['SibSp', 'Parch', 'FamileSize'], axis=1)
test_df = test_df.drop(['SibSp', 'Parch', 'FamileSize'], axis=1)
combine = [train_df, test_df]
train_df.head(6)
```

```python
fre_port = train_df.Embarked.dropna().mode()[0]
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].fillna(fre_port)
```

```python
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].map({'s':0, 'c':1, 'Q':2}).astype(int)
train_df.head()
```
